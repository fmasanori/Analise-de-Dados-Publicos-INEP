### 1- Importando pandas e o arquivo csv

```python
import pandas as pd
filename = 'DM_CURSO.csv'
```

### 2- Colocando os dados no dataframe

```python
df = pd.read_csv(filename, encoding='iso-8859-1', header=None, delimiter='|')
df
```

### 3- Selecionando as colunas necessárias para análise
###     Coluna 14 --> Nome do Curso
###     Coluna 91 --> Quantidade de Matrículas

```python
df1 = df[[14, 91]]
df1
```

### 4- Renomeando as colunas

```python
df1.rename(columns={14: 'NomeDoCurso', 91: 'MatriculasNoCurso'}, inplace=True)
df1
```

```python
df1 = df1.drop([0])
df1
```

### 5- Formatando as colunas para facilitar a análise

```python
a = df1
a = df1[a.MatriculasNoCurso.notnull()]
```

```python
a2 = a
a2 = a2[a2.MatriculasNoCurso != '']
```

```python
a3 = a2
a3 = a2.drop_duplicates('NomeDoCurso')
```

```python
a4 = a3
a4['MatriculasNoCurso'] = a4['MatriculasNoCurso'].astype(float)
```

### 6- Com as colunas formatadas é selecionado o Top10 de cursos com mais matrículas

```python
a5 = a4.sort_values(['MatriculasNoCurso'],ascending = False).head(10) 
```

```python
a5
```

### 7- Gerando o Gráfico Final

```python
a5.plot.barh(x='NomeDoCurso', y='MatriculasNoCurso')
```
